```python
import pandas as pd
import numpy as np
import copy
import QSTK.qstkutil.qsdateutil as du
import datetime as dt
import QSTK.qstkutil.DataAccess as da
import QSTK.qstkutil.tsutil as tsu
import QSTK.qstkstudy.EventProfiler as ep
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ls_symbols = ["AAPL","MSFT","GOOG","IBM"]
dt_start = dt.datetime(2010, 1, 1)
dt_end = dt.datetime(2010, 12, 31)
timeofday=dt.timedelta(hours=16)
ldt_timestamps = du.getNYSEdays(dt_start, dt_end, timeofday)
ls_keys = ['open', 'high', 'low', 'close', 'volume', 'actual_close']
dataobj = da.DataAccess('Yahoo', cachestalltime=0)
ldf_data = dataobj.get_data(ldt_timestamps, ls_symbols, ls_keys)
d_data = dict(zip(ls_keys, ldf_data))
```

```python
for s_key in ls_keys:
    d_data[s_key] = d_data[s_key].fillna(method='ffill')
    d_data[s_key] = d_data[s_key].fillna(method='bfill')
    d_data[s_key] = d_data[s_key].fillna(1.0)
```

```python
d_data['rolling_avg'] = pd.rolling_mean(d_data['close'],20,min_periods=20)
```

```python
d_data['rolling_std'] = pd.rolling_std(d_data['close'],20,min_periods=20)
```

```python
Bollinger_val = (d_data['close'] - d_data['rolling_avg'])/d_data['rolling_std'] 
```
